```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.misc import imsave
from __future__ import unicode_literals
import pandas as pd
%matplotlib inline
```

```python
def config_axis(ax):
    ax.grid(True)
    ax.legend(loc = "best", fontsize = 28)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(24)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(24)
    ax.yaxis.label.set_fontsize(28)
    ax.xaxis.label.set_fontsize(28)
```

```python
def plot_frame_values(frames, action_values, xname, yname, ann_ids, fig_name = None):
    fig = plt.figure(figsize=(16, 12))
    ax = fig.add_subplot(111)
    ax.grid(True)
    ax.plot(frames, action_values, linewidth = 5, color = '#2A6EA6')
    
    for x, i in enumerate(ann_ids):
        j = action_values[i - frames[0]]
        ax.annotate(chr(ord('A') + x), xy=(i, j), xycoords='data', xytext=(i - 2., j + 0.1),
                    arrowprops=dict(facecolor='black', shrink=0.15, frac=0.3), fontsize = 28)
    
    ax.set_xlabel(xname, fontsize = 28)
    ax.set_ylabel(yname, fontsize = 28)
    config_axis(ax)
        
    fig.tight_layout()
#     plt.suptitle(plotname, fontsize = 24, fontweight = 'bold')
    if fig_name is not None:
        plt.savefig(fig_name, dpi=100)
    plt.show()
```

# Plot average Q-values and rewards

```python
df = pd.DataFrame.from_csv('./values.csv')
df.head()
```

```python
frames = df.index
action_values = df.action_values.values
start = 170
end = 211
frames = frames[start:end]
action_values = action_values[start:end]
print frames.shape
print action_values.shape
fig_name = 'dqn_seaquest_frames.png'
plot_frame_values(frames, action_values, 'Frame', 'Giá trị hành động lớn nhất', [8182, 8186, 8188, 8191], fig_name)
```
