```python
import numpy as np
import matplotlib.pyplot as plt
from __future__ import unicode_literals
%matplotlib inline
```

```python
np.random.seed(24315131)

# example data
n = 10
x = np.linspace(1, n, n)
y = np.random.randn(len(x))
yerr = abs(np.random.rand(len(x)) / 2) + 0.3

plt.figure()
plt.errorbar(x, y, yerr=yerr, fmt='o')
# plt.title("Đồ thị giá trị hành động và sai số tương ứng", fontsize = 16)
plt.axis([min(x) - 1, max(x) + 1, min(y - yerr) - 0.1, max(y + yerr) + 0.1])
plt.xlabel('Hành động', fontsize = 14)
plt.xticks(np.linspace(1, n, n))
plt.ylabel('Giá trị hành động', fontsize = 14)

ymax = max(y)
plt.plot([min(x) - 1, max(x) + 1], [ymax, ymax], linewidth = 2)
plt.tight_layout()
plt.savefig('q_learning.png', dpi = 300)
plt.show()
```

```python
np.random.seed(24315131)

# example data
n = 10
x = np.linspace(1, n, n)
y = np.random.randn(len(x))
yerr = abs(np.random.rand(len(x)) / 2) + 0.3

plt.figure(figsize = (16, 8))
plt.subplot(1, 2, 1)
plt.errorbar(x, y, yerr=yerr, fmt='o')
plt.title("Bộ tham số $\\theta$", fontsize = 24)
plt.axis([min(x) - 1, max(x) + 1, min(y - yerr) - 0.1, max(y + yerr) + 0.1])
plt.xlabel('Hành động', fontsize = 24)
plt.xticks(np.linspace(1, n, n))
plt.ylabel('Giá trị hành động', fontsize = 24)
for tick in plt.gca().xaxis.get_major_ticks():
    tick.label.set_fontsize(20)
for tick in plt.gca().yaxis.get_major_ticks():
    tick.label.set_fontsize(20)
ymax = max(y)
plt.gcf().tight_layout()
plt.plot([min(x) - 1, max(x) + 1], [ymax, ymax], linewidth = 2)

n = 10
x = np.linspace(1, n, n)
y = np.random.randn(len(x))
yerr = abs(np.random.rand(len(x)) / 2) + 0.3
plt.subplot(1, 2, 2)
plt.errorbar(x, y, yerr=yerr, fmt='o')
plt.title("Bộ tham số $\\theta^{-}$", fontsize = 24)
plt.axis([min(x) - 1, max(x) + 1, min(y - yerr) - 0.1, max(y + yerr) + 0.1])
plt.xlabel('Hành động', fontsize = 24)
plt.xticks(np.linspace(1, n, n))
plt.ylabel('Giá trị hành động', fontsize = 24)

for tick in plt.gca().xaxis.get_major_ticks():
    tick.label.set_fontsize(20)
for tick in plt.gca().yaxis.get_major_ticks():
    tick.label.set_fontsize(20)
    
ymax = max(y)
plt.gcf().tight_layout()
plt.plot([min(x) - 1, max(x) + 1], [ymax, ymax], linewidth = 2)
plt.savefig('double_q_learning.png', dpi = 200)
plt.show()
```
